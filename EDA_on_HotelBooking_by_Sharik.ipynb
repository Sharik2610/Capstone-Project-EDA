## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#creating a copy of dataframe
hotel_df = df.copy()
````

````python
#dropping the duplicate rows
hotel_df= hotel_df.drop_duplicates()
hotel_df
````

````python
#data frame after filtering duplicates
print(f'we have total {hotel_df.shape[0]} rows and {hotel_df.shape[1]} columns.')
````

*  It is better to drop the column 'company' altogether since the number of missing values is extremely high compared to the number of rows.



````python
# droping the column which has maximum no.of null values
hotel_df.drop(['company'], axis = 1, inplace=True)
````

````python
hotel_df.head().T
````

````python
# converting the datatype of reservation_status_date to date

hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'], format='%Y-%m-%d')
````

````python
# Adding total stays in nights

hotel_df['total_stays_in_nights'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']
````

````python
# Adding revenue

hotel_df['revenue'] = hotel_df['total_stays_in_nights'] * hotel_df['adr']
````

````python
# Adding total number of guests as column

hotel_df['total_guests'] = hotel_df['adults'] + hotel_df['children'] + hotel_df['babies']
````

````python
# Adding column which shows whether reserved room type is the same as assigned room type

hotel_df['is_reserved_room_type_assigned'] = hotel_df.apply(lambda x: 'same room type assigned' if x['reserved_room_type'] == x['assigned_room_type'] 
                                                                                                   else 'different room type assigned', axis=1)
````

````python
# changing the bool data from int to string for easy representation

hotel_df['is_canceled'] = hotel_df['is_canceled'].replace([1, 0], ['cancelled', 'not cancelled'])
hotel_df['is_repeated_guest'] = hotel_df['is_repeated_guest'].replace([1, 0], ['repeated guest', 'not repeated guest'])
````

````python
hotel_df.head().T
````

````python
file_path2= '/content/wikipedia-iso-country-codes.csv'
country_code= pd.read_csv(file_path2)
````

````python
country_code
````

````python
#
country_code.rename(columns = {'English short name lower case' :'Country'},inplace = True)
country_code
````

````python
country_code.drop(['Alpha-2 code','Numeric code',	'ISO 3166-2'],axis=1,inplace=True )
````

````python
country_code['country'] = country_code['Alpha-3 code']
country_code
````

````python
hotel_df_new = pd.merge(hotel_df, country_code)

hotel_df_new.drop([	'Alpha-3 code'],axis=1,inplace=True)

hotel_df_new
````

````python
#Let's look at the country-wise guests that both the hotels in total have received.


country_wise_guests = hotel_df_new[hotel_df_new['is_canceled'] == 'not cancelled']['Country'].value_counts().reset_index().head(10)
country_wise_guests.columns = ['Country', 'No of guests']
country_wise_guests

plt.figure(figsize=(12,6))

plt.title('Top 10 countries from where these hotels are receiving guests')
sns.barplot(x=country_wise_guests['Country'], y=country_wise_guests['No of guests']);

````
